```python
df["sqft_living"] = pd.to_numeric(df["sqft_living"])
mean = df["sqft_living"].mean()
median =df["sqft_living"].median()
mode = df["sqft_living"].mode()
print(mean , median, mode)
```

```python
df.sqft_living.value_counts().nlargest(30).plot(kind='bar', figsize=(10,6))
plt.title("Number of rooms by sqft_living")
plt.ylabel('Number of rooms')
plt.xlabel('sqft_living of the room')
```

```python
sns.boxplot(x="sqft_living",data=df)
plt.show()
```

```python
plt.scatter(df["sqft_living"], df["sqft_lot"])
plt.title("Scatter Plot for sqft_lot vs sqft_living")
plt.xlabel("sqft_living")
plt.ylabel("sqft_lot")
```

```python
sns.boxplot(x="bedrooms",y="price",data=df)
plt.show()
```

```python
sns.pairplot(df,vars = ['condition', 'sqft_above','sqft_basement'], kind="reg")
plt.show()
```

```python
sns.set(style="ticks", color_codes=True)
sns.pairplot(df,height=2,vars = ['condition', 'sqft_above','sqft_basement'], hue="waterfront")
plt.show()
```

```python
from scipy import stats

corr = stats.pearsonr(df["sqft_above"], df["sqft_basement"])
print("p-value:\t", corr[1])
print("cor:\t\t", corr[0])
```

```python
correlation = df.corr(method='pearson', numeric_only=True)
correlation
```

```python
sns.heatmap(correlation,xticklabels=correlation.columns,
            yticklabels=correlation.columns)
```

Module - 3 (Time Series Analysis)

```python
df['date'] = pd.to_datetime(df['date'])
df.dtypes
```

```python
df = df.set_index('date')
df.tail(3)
```

```python
df.index
```

```python
df['Year'] = df.index.year
df['Month'] = df.index.month
df['Weekday Name'] = df.index.day_name()
```

```python
df.sample(5, random_state=0)
```

```python
df.loc['2014-05-20']
```

```python
sns.set(rc={'figure.figsize':(11, 4)}) 
plt.rcParams['figure.figsize'] = (8,5)
plt.rcParams['figure.dpi'] = 150
```

```python
df['price'].plot(linewidth=0.25)
```

```python
cols_to_plot = ['sqft_lot', 'sqft_above', 'sqft_basement']
axes = df[cols_to_plot].plot(marker='*', alpha=0.5,
linestyle='None',figsize=(14, 6), subplots=True) 
for ax in axes:
  ax.set_ylabel('Sqft')
```

```python
ax = df.loc['2014', 'price'].plot()
ax.set_ylabel('Price')
```

```python
ax = df.loc['2014-12', 'price'].plot(marker='o', linestyle='-')
ax.set_ylabel('Price')
```

```python
fig, axes = plt.subplots(3, 1, figsize=(8, 7), sharex=True)
for name, ax in zip(['sqft_lot', 'sqft_above', 'sqft_basement'], axes):
  sns.boxplot(data=df, x='Month', y=name, ax=ax)
  ax.set_ylabel('Sqft')
  ax.set_title(name)
  if ax != axes[-1]:
    ax.set_xlabel('')
```
